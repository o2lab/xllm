```python
import torch
with torch.no_grad():
    torch.cuda.empty_cache()
```

```python
from transformers import  AutoTokenizer, AutoModelForCausalLM, QuantoConfig

model_id = "mistralai/Mixtral-8x7B-Instruct-v0.1"
quantization_config = QuantoConfig(weights="int4", modules_to_not_convert=["model.layers.8.w1", "model.layers.7.w1", "model.layers.6.w1", "model.layers.5.w1"])
model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config, device_map='auto')
tokenizer = AutoTokenizer.from_pretrained(model_id)
```
